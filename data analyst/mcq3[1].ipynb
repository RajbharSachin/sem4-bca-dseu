```python
# q1 

import pandas as pd
import matplotlib.pyplot as plt

data={
    'Program':['Btech','BCA','Diploma'],
    'No_of_Course':[40,10,36]
}

df=pd.DataFrame(data)
df.to_csv('Dseu_technicl_course.csv')
df=pd.read_csv('Dseu_technicl_course.csv')

plt.bar(df['Program'],df['No_of_Course'])
plt.xlabel('Program')
plt.ylabel('Number of courses')
plt.title('Number of courses provided by DSEU')
plt.show()
```

```python
# q2 
#  screen time dta 

import pandas as pd
import matplotlib.pyplot as plt

# Assume data for screen time of students (in minutes)
boys_screen_time = [120, 90, 160, 180, 135, 165, 110, 200]
girls_screen_time = [150, 80, 120, 150, 130, 160, 90, 180]

# Create a DataFrame for boys and girls
data = {'Boys': boys_screen_time, 'Girls': girls_screen_time}
df = pd.DataFrame(data)

# Create a box plot
plt.boxplot([boys_screen_time, girls_screen_time], labels=['Boys', 'Girls'])
plt.xlabel('Gender')
plt.ylabel('Screen Time (minutes)')
plt.title('Screen Time of Students')

# Display the box plot
plt.show()
```
